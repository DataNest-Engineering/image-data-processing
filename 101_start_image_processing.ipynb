```python
cv2.imwrite("tests/resized_img1.jpg", cv2.resize(img, (500, 100)))
```

```python
img_resized = cv2.resize(img, (500, 500))
```

```python
# Flip the image: 1 = horizontal, 0 = vertical, -1 = both
img_flipped_horizontal = cv2.flip(img_resized, 1)
cv2.imwrite("tests/flipped_horizontal.jpg", img_flipped_horizontal)
```

```python
cv2.imwrite("tests/stacked.jpg",np.hstack((img_resized, img_flipped_horizontal)))
```

```python
img_flipped_vertical = cv2.flip(img_resized, 0)
cv2.imwrite("tests/flipped_vertical.jpg", img_flipped_vertical)
```

```python
cv2.imwrite("tests/stacked_vertical.jpg", np.hstack((img_resized, img_flipped_vertical)))
```

```python
img_flipped_both = cv2.flip(img_resized, -1)
cv2.imwrite("tests/flipped_both.jpg", img_flipped_both)
```

```python
cv2.imwrite("tests/stacked_both.jpg", np.hstack((img_resized, img_flipped_both)))
```

```python
# 4. DRAWING ON THE IMAGE
# Format: (image, coordinates, color_BGR, thickness)
# Note: (0,0) is the top-left corner of the image

```

```python
img_flipped_horizontal
```

```python
# Draw a Green Line (BGR: 0, 255, 0)
cv2.line(img_flipped_horizontal, (0, 0), (200, 200), (0, 255, 0), 5)
```

```python

cv2.imwrite("tests/line.jpg", cv2.line(img_flipped_horizontal, (0, 0), (200, 200), (255, 50, 0), 5))
```

```python

# Draw a Red Rectangle (Top-Left, Bottom-Right)
cv2.imwrite("tests/rectangle.jpg", cv2.rectangle(img_flipped_horizontal, (300, 50), (450, 150), (0, 0, 255), 3))
```

```python

# Draw a Blue Filled Circle (Center, Radius)
# Using -1 for thickness fills the shape
cv2.imwrite("tests/circle.jpg", cv2.circle(img_flipped_horizontal, (100, 400), 50, (255, 0, 0), -1))
```

```python

# Add White Text
cv2.imwrite("tests/white_text.jpg", cv2.putText(img_flipped_horizontal, "OpenCV Lesson 1", (150, 480), 
            cv2.FONT_HERSHEY_COMPLEX, 1, (255, 255, 255), 2))
```

```python

# 5. STACKING AND DISPLAYING
# To show images side-by-side, they must have the same number of channels.
# Since img_gray has 1 channel, we convert it back to 3 channels to stack with RGB.
gray_3_channel = cv2.cvtColor(img_gray, cv2.COLOR_GRAY2BGR)
gray_resized = cv2.resize(gray_3_channel, (500, 500))
```

```python
# Horizontal stack: Original (resized) next to the Edited version
hor_stack = np.hstack((gray_resized, img_flipped_horizontal))
```

```python
cv2.imshow("Left: Gray | Right: Edited & Flipped", hor_stack)
```

```python
cv2.imwrite("tests/stacked.jpg", np.hstack((gray_resized, img_flipped_horizontal)))
```

```python
cv2.imwrite("tests/vstacked.jpg", np.vstack((gray_resized, img_flipped_horizontal)))
```
